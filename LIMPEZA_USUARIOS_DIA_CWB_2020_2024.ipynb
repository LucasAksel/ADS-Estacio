```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/CÓDIGOS DO AMANHÃ/PROJETO/Usuários_dia_CWB_2020_2024.csv')

df.info()
```

```python
df.head()
```

```python
#Trocar nomes e criar coluna totais

df.rename(columns={'TOTAL S/ ISENTOS': 'PAGANTES', 'Passageiros Isentos': 'ISENTOS'}, inplace=True)

df['TOTAL'] = df['PAGANTES'] + df['ISENTOS']

display(df.head())
```

```python
#Correção de formato de data
df_correcao_datas = df.copy()

    #Usa o formato escrito da data para transformar em int
df_correcao_datas['DATA'] = pd.to_datetime(df_correcao_datas['DATA'], format='%d de %b. de %Y', errors='coerce')

df_correcao_datas.info()
```

```python
#Validar a correção
df_correcao_datas.head()

df_correcao_datas.describe()
```

```python
df_mensal = df_correcao_datas.copy()

# .to_period faz com que o período selecionado (que no caso foi mês) será aglomerado, estando o dado original em formato de data legível
df_mensal['Mes'] = df_mensal['DATA'].dt.to_period('M')

# Agrupar e somar as colunas de TOTAl e de isentos
df_mensal_agregado = df_mensal.groupby('Mes')[['PAGANTES', 'ISENTOS', 'TOTAL']].sum().reset_index()

display(df_mensal_agregado)
```
